## 4. Gravity Example

Let's look at the script for a particular gravity study and expand from there. This case study can be seen in this link: http://docs.simpeg.xyz/content/examples/04-grav/plot_inversion_linear.html#sphx-glr-content-examples-04-grav-plot-inversion-linear-py <br>

### <center> START! </center>

**Installing packages** <br>
First, let's import the necessary packages for this problem


````python
import numpy as np
import matplotlib.pyplot as plt

#This one contains all the main packages you need from SimPEG. 
from SimPEG import Mesh, Utils, Maps, Regularization, DataMisfit, Optimization, InvProblem, Directives, Inversion, PF
````

**Create our mesh**<br>
Second, let's create our mesh for this problem. Seems like they used a discretization of dx = 5. 

````python
dx = 5.
incr = 1.3
ymax = 15.

hxind = [(dx, dx, -incr), (dx, 15), (dx, dx, incr)]
hyind = [(dx, dx, -incr), (dx, 15), (dx, dx, incr)]
hzind = [(dx, dx, -incr), (dx, 7), (3.5, 1), (2, dx)]

mesh = Mesh.TensorMesh([hxind, hyind, hzind], 'CCC')
#C means center in the first axis, likewise CC means it centers in second axis, and CCC is center on all axis

print(mesh)
mesh.plotGrid()
````

Then let's find the center index (need this for integral approximation I assume)


````python
midx = int(mesh.nCx/2) #for this problem, our nCx and nCy are 25, because we have 5+15+5 points every h
midy = int(mesh.nCy/2)
````

**Gaussian Topo**<br>
Need to figure this out, but it's like, create the topography for our survey? <br>
The topography would be: <br>
$$z = -exp({\frac{x^2 + y^2}{75^2}}) + 53.64 $$
Here's how you create the topography:

````python
[xx, yy] = np.meshgrid(mesh.vectorNx, mesh.vectorNy)
zz = -np.exp((xx**2+yy**2)/75**2) + mesh.vectorNz[-1]

### This is usually where you load your topography file, but let's make it synthetic for now
topo = np.c_[Utils.mkvc(xx), Utils.mkvc(yy),  Utils.mkvc(zz)]

````

- *Utils.mkvc()* is used to create a vector with number of dimension specified. <br>
It is located in the DISCRETIZE library <br>
- *np.c_* is another for of array concatenation 

For going from topo to active cell:

````python
actv = Utils.surface2ind_topo(mesh,topo,'N') #gets active indices from topography 
#input: mesh to which the function discretize the topography
#topo as in t\he topography data
#N: nodes, CC: center

actv = np.asarray([inds for inds, elem in enumerate(actv,1) if elem], dtype=int)-1
#creates an array for the activated cells in the previous line
````

Create a map to go from reduced space to full:

````python
actvMap = Maps.InjectActiveCells(mesh, actv, -100)

nC = len(actv)
print("number of active cells:",nC)
````

I assume the above line of code injects the active cell to the mesh so we would have a fuller matrix in our model? (need ask)

**Survey time!**<br>
Now let's make the survey and get those data. <br>
First, let's create an array of observation points, then we'll create the survey.
In the x-axis, you have 20 points, and so is in the y-axis. Therefore we have 400 survey stations

````python
# 1. array of observation points
xr = np.linspace(-30.,30.,20)
yr = np.linspace(-30.,30.,20)
X,Y = np.meshgrid(xr,yr)

#observation points 5m above the topo
Z = -np.exp((X**2 + Y**2) / 75**2) + mesh.vectorNz[-1] + 5 #the code has it as 0.1 m

#2. Survey design:
rxloc = np.c_[Utils.mkvc(X.T), Utils.mkvc(Y.T), Utils.mkvc(Z.T)]
rxLoc = PF.BaseGrav.RxObs(rxloc)

plt.plot(rxloc,'.')
plt.title('Rx-Location')

#source field:
srcField = PF.BaseGrav.SrcField([rxLoc]) #defines the inducing field 
survey = PF.BaseGrav.LinearSurvey(srcField) #project field to data space
````

**Create a Susceptibility Model!** <br>
This will be the target to our survey!

````python
#initialize the model array with zeros first, and then fill it in as a block of susceptibility afterwards
model = np.zeros((mesh.nCx, mesh.nCy,mesh.nCz))
model[(midx-5):(midx-1), (midy-2):(midy+2), -10:-6] = 0.75 #g/cm^3
model[(midx+1):(midx+5), (midy-2):(midy+2), -10:-6] = -0.75
print("model dimension:", model.shape)


model = Utils.mkvc(model) #create a model vector with "model" as the dimension 25x25x18
model = model[actv]  #model is only the active cell ones

#map for full:
actvMap = Maps.InjectActiveCells(mesh, actv, -100)

#reduced identity map (what???)
idenMap = Maps.IdentityMap(nP=nC) #nP is the number of parameters the mapping accept
#I guess it changes the physical property into the model?

````

** Forward Modelling**<br>
Let's find the FM operator:

````python
prob = PF.Gravity.GravityIntegral(mesh, rhoMap=idenMap, actInd=actv)
#Accepts model(mesh), density(idenMap), and map of density-Inversion model(actv), spits out the forward model
````

````python
##########################################
#......   WARNING!!
######this cell can only be run once######
######if not, it'll break#################

survey.pair(prob) #pairs survey with forward model 

#Forward operator d=Gm can be coded like this:
d = prob.fields(model)
````

This means the G matrix in this case is prob.fields(), which we calculated using GravityIntegral


**Let's make some noise!**<br>
Adding some random Gaussian noise (1nT)


````python
data = d+np.random.randn(len(d))*1e-3 
wd = np.ones(len(data))*1e-3 #WEIIIGHTS flat uncertainties in our measurement


plt.plot(data, label="noise")
plt.plot(d, label="noiseless")
plt.legend()
plt.title("Just the Data Points")

````

Now that we have the noisy survey, let's store that in our survey:

````python
survey.dobs = data #observed data
survey.std = wd #standard deviation/weights on the data
survey.mtrue = model #true model in the survey

#Since Forward modelling can have errors too, then we shall put more WEIIGHTS from that:
rxloc = survey.srcField.rxList[0].locs
wr = np.sum(prob.G**2, axis=0)**0.5
wr = wr/np.max(wr) #this has like, 10085 elements, damn
````

**Regularization**<br>
Using Regularization.Sparse, we can use this to find the regularization: <br>
 $$R(m) = \frac{1}{2}\mathbf{(m-m_\text{ref})^\top W^\top R^\top R
 |          W(m-m_\text{ref})}$$

````python
reg = Regularization.Sparse(mesh, indActive=actv, mapping=idenMap)
reg.cell_weights = wr
reg.norms = np.c_[1,0,0,0] #sparse regularization
````

**Find Misfit Function**<br>
Seems like we're using l2 approximation, using weights from the data. The math behind DataMisfit.l2_DataMisfit is:
$$ \mu_\text{data} = {1\over 2}\left|
 |      \mathbf{W}_d (\mathbf{d}_\text{pred} -
 |      \mathbf{d}_\text{obs}) \right|_2^2$$
 Where $\mu$ is misfit

````python
dmis = DataMisfit.l2_DataMisfit(survey)
dmis.W = Utils.sdiag(1/wd)
````

** Inversion Time!!**<br>
First let's add directives to the inversion (whatever that means :thinking_face:

````python


#The following line minimizes a general class for derivative based optimization
# There's 5 different methods for optimization, but we're using ProjectedGNCG now
#TODO: document the other 4 methods
opt = Optimization.ProjectedGNCG(maxIter=100, lower=-1., upper=1.,
                                     maxIterLS=20, maxIterCG=10,
                                     tolCG=1e-3) 

#Actual inverse problem
invProb = InvProblem.BaseInvProblem(dmis, reg, opt)

#Find Beta
betaest = Directives.BetaEstimate_ByEig()
````

**Something something Norms and Jacobian??? Inversion directives**<br>
This just proves that you need to study more about Jacobian

````python
##########################################
#......   WARNING!!
######this cell can only be run once######
######if not, it'll break#################

IRLS = Directives.Update_IRLS(f_min_change=1e-4, maxIRLSiter=30, coolEpsFact = 1.5, beta_tol=1e-1, )

saveDict = Directives.SaveOutputEveryIteration(save_txt=False)
update_Jacobi = Directives.UpdatePreconditioner() #creates a Jacobi preconditioner
inv = Inversion.BaseInversion(invProb, directiveList=[IRLS, betaest, update_Jacobi, saveDict])
````

<center>***NOW YOU DO THE INVERSION!***</center>

````python
m0 = np.ones(nC)*1e-4 #starting model
mrec = inv.run(m0)
````

**PLOTTING**

**Data**

````python
 # Here is the recovered susceptibility model

#specifies which plane are we looking at
ypanel = midx
zpanel = -7

#l2model that we found is dotted with the active map to get the model and also regularized to not have erroneous data
m_l2 = actvMap * invProb.l2model
m_l2[m_l2 == -100] = np.nan

#Truncated solution? with mrec times the active map?
m_lp = actvMap * mrec
m_lp[m_lp == -100] = np.nan

#true model
m_true = actvMap * model
m_true[m_true == -100] = np.nan

vmin, vmax = mrec.min(), mrec.max()
````
